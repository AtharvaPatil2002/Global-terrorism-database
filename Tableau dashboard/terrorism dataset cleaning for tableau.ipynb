```python
df.rename(columns={'iyear' : 'Year', "imonth": "Month", "iday":"Day", "country": "Country_no", 
                   "country_txt":"Country", "region_txt":"Region", "city": "City", 
                   "propvalue": "property_value", "attacktype1_txt":"Attack_type", 
                   "targtype1_txt": "Target_type", "gname": "group_name", "targsubtype1_txt": "Target_sub_type", 
                   "target1": "Target", "nkill": "no_of_kills", "weaptype1_txt": "Weapon_type", 
                   "weapdetail": "Weapon_detail"}, inplace = True)
```

```python
df["City"].fillna("Unknown", inplace=True)
df["Target"].fillna("Unknown", inplace=True)
df["property_value"].fillna(0, inplace=True) 
df["no_of_kills"].fillna(0, inplace=True)
```
